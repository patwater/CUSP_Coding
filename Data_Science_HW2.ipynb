# Problem 1

```python
import numpy as np

normal = np.random.normal(0,1,10)
print "The mean is "+str(normal.mean())
print "The std dev is "+str(normal.std())
print "The variance is "+str(normal.std()**2)
```

```python
normal = np.random.normal(0,1,1000)
print "The mean is "+str(normal.mean())
print "The std dev is "+str(normal.std())
print "The variance is "+str(normal.std()**2)
```

```python
normal = np.random.normal(0,1,10**6)
print "The mean is "+str(normal.mean())
print "The std dev is "+str(normal.std())
print "The variance is "+str(normal.std()**2)
```

As the sample size increases, the mean, standard deviation and variance tend towards their expected value (i.e. mean = 0, standard deviation and varinace = 1).  These examples illustrate the law of large numbers.

```python
from collections import Counter

with open('x_random_1000_ints(0,9).txt') as raw_file:
    X = []
    for line in raw_file:
        X.append(int(line))

with open('y_random_1000_ints(0,9).txt') as raw_file:
    Y = []
    for line in raw_file:
        Y.append(int(line))   

coord = []
for i in xrange(len(X)):
    coord.append(tuple((X[i],Y[i])))

coord_count = Counter(coord)
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)
model = sm.OLS(Y,X)
result = model.fit()

print result.summary()
```

Our t statistic here is .832, yielding a p value of .406.  This is below our threshold of 2 in absolute value and we fail to reject the null hypothesis that there is no relationship between X and Y.  (If we were feeling Bayesian, we might note that our prior would be that there's no relationship between X and Y since both are completely random.)

```python
# Just for fun, seeing what this looks like

from matplotlib import pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(X,Y, 'o', label='data')
ax.plot(X, result.fittedvalues, label="OLS")
plt.show()
```

```python
color_list = ['black','blue','orange','red','purple']
parameters_list =[]

for i in xrange(5):
    model = sm.OLS(yi[i], xi[i])
    result = model.fit()
    parameters_list.append(result.params)
    print result.summary()
    
    fig, ax = plt.subplots(figsize=(8,6))
    ax.plot(xi[i], yi[i], 'o', color = color_list[i])
    ax.plot(xi[i], result.fittedvalues, label="OLS")
    plt.show()
```
